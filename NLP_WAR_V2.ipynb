```

xl = pd.ExcelFile("WAR Excel.xlsx")
#xl.sheet_names
df = xl.parse("WAR 2")
df.head()
```

```
#import en_core_web_sm
nlp = en_core_web_sm.load()   

GPE_submission_dict = {}
LOC_submission_dict = {}
PERSON_submission_dict = {}
ORG_submission_dict = {}
WORK_OF_ART_submission_dict = {}
EVENT_submission_dict = {}
LAW_submission_dict = {}
FAC_submission_dict = {}
LANGUAGE_submission_dict = {}

for a_details, a_date in zip(df['Activity Details'], df['Activity Date']):
  #lem_detaila = stemer_lemmatizer(a_details)

  doc = nlp(a_details.lower())

  for ent in doc.ents:
      #print(ent.text, ent.start_char, ent.end_char, ent.label_) 
      if ent.label_ == 'GPE':
        if ent.text.strip() == 'us' or ent.text.strip()  == 'u.s.' or ent.text.strip()  == 'the united states' or ent.text.strip()  == 'america':
          GPE_submission_dict[a_date] = 'United States'
        else:
          GPE_submission_dict[a_date] = ent.text.strip().capitalize()
      if ent.label_ == 'LOC':
        LOC_submission_dict[a_date] = ent.text.strip().capitalize()
      if ent.label_ == 'PERSON':
        PERSON_submission_dict[a_date]  = ent.text.strip().capitalize()
      if ent.label_ == 'WORK_OF_ART':
        WORK_OF_ART_submission_dict[a_date]  = ent.text.strip().capitalize()    
      if ent.label_ == 'ORG':
        ORG_submission_dict[a_date]  = ent.text.strip().capitalize() 
      if ent.label_ == 'EVENT':
        EVENT_submission_dict[a_date]  = ent.text.strip().capitalize() 
      if ent.label_ == 'LAW':
        LAW_submission_dict[a_date]  = ent.text.strip().capitalize() 
      if ent.label_ == 'FAC':
        FAC_submission_dict[a_date]  = ent.text.strip().capitalize() 
      if ent.label_ == 'LANGUAGE':
        LANGUAGE_submission_dict[a_date]  = ent.text.strip().capitalize() 

GPE_DF = pd.DataFrame(list(GPE_submission_dict.items()),columns = ['Date','GPE'])
GPE_DF.to_csv('GPE_DF.csv', index=False)

LOC_DF = pd.DataFrame(list(LOC_submission_dict.items()),columns = ['Date','LOC'])
LOC_DF.to_csv('LOC_DF.csv', index=False)

PERSON_DF = pd.DataFrame(list(PERSON_submission_dict.items()),columns = ['Date','PERSON'])
PERSON_DF.to_csv('PERSON_DF.csv', index=False)

ORG_DF = pd.DataFrame(list(ORG_submission_dict.items()),columns = ['Date','ORG'])
ORG_DF.to_csv('ORG_DF.csv', index=False)

WORK_OF_ART_DF = pd.DataFrame(list(WORK_OF_ART_submission_dict.items()),columns = ['Date','WORK_OF_ART'])
WORK_OF_ART_DF.to_csv('WORK_OF_ART_DF.csv', index=False)

EVENT_DF = pd.DataFrame(list(EVENT_submission_dict.items()),columns = ['Date','EVENT'])
EVENT_DF.to_csv('EVENT_DF.csv', index=False)

LAW_DF = pd.DataFrame(list(LAW_submission_dict.items()),columns = ['Date','LAW'])
LAW_DF.to_csv('LAW_DF.csv', index=False)


```
